```python
import pandas as pd
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt

# Suponha que você já tenha um dataframe chamado 'df' com duas colunas chamadas 'coluna1' e 'coluna2'.
df = pd.read_csv('/home/gsarini/Documentos/Doc/paper_regfilter/bra_006/comparacao_splvsdda/bra006_dda_vs_spl_shared_compounds.csv')

# Calcula a média e o desvio padrão das duas colunas
media_coluna1 = df['SPL'].mean()
std_coluna1 = df['SPL'].std()

media_coluna2 = df['DDA'].mean()
std_coluna2 = df['DDA'].std()

# Realiza o teste t de Student para comparar as médias
stat, p_value = ttest_ind(df['DDA'], df['SPL'])

# Exibe as médias, desvio padrão e o resultado do teste estatístico
print(f'Média da coluna 1: {media_coluna1}, Desvio Padrão: {std_coluna1}')
print(f'Média da coluna 2: {media_coluna2}, Desvio Padrão: {std_coluna2}')
print(f'Resultado do teste t de Student: Estatística={stat}, Valor p={p_value}')

# Cria um gráfico de barras com as médias
colunas = ['SPL', 'DDA', 'Resultado Teste T']
valores = [media_coluna1, media_coluna2, stat]

plt.bar(colunas, valores)

# Adiciona as informações de média e desvio padrão no gráfico
plt.text(0, media_coluna1, f'Média: {media_coluna1}\nDesvio Padrão: {std_coluna1}', ha='center', va='bottom')
plt.text(1, media_coluna2, f'Média: {media_coluna2}\nDesvio Padrão: {std_coluna2}', ha='center', va='bottom')

plt.ylabel('Valores')
plt.title('Médias das Colunas e Resultado do Teste T')
plt.show()
```

```python
import pandas as pd
from scipy.stats import ttest_ind
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt

# Suponha que você já tenha um dataframe chamado 'df' com duas colunas chamadas 'coluna1' e 'coluna2'.

# Calcula a média das duas colunas
media_coluna1 = df['SPL'].mean()
media_coluna2 = df['DDA'].mean()

# Realiza o teste t de Student para comparar as médias
stat, p_value = stats.ttest_ind(df['SPL'], df['DDA'], equal_var=True)

# Exibe as médias e o resultado do teste estatístico
print(f'Média quantidade de picos compartilhados SPL: {media_coluna1}')
print(f'Média quantidade de picos compartilhados DDA: {media_coluna2}')
print(f'Resultado do teste t de Student: Estatística={stat}, Valor p={p_value}')

# Cria um gráfico de barras usando a biblioteca seaborn
data = pd.DataFrame({'Média': [media_coluna1, media_coluna2],
                     '': ['SPL', 'DDA']})

# Calcula o desvio padrão das colunas
std_coluna1 = df['SPL'].std()
std_coluna2 = df['DDA'].std()

# Cria o gráfico com barras e desvio padrão usando seaborn
plt.figure(figsize=(8, 6))
sns.barplot(x='', y='Média', data=data, yerr=[std_coluna1, std_coluna2])
plt.ylabel('Shared Peaks')
plt.title('Médias das Colunas e Resultado do Teste T com Desvio Padrão')
plt.show()
```
